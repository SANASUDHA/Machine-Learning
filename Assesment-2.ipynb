```python
import pandas as pd
```

```python
iris = pd.read_csv('Iris.csv')
iris.head()
```

```python
iris.Species.value_counts()
```

```python
print("Feature : Sepal Length")
print('max:',iris.SepalLengthCm.max())
print('min:',iris.SepalLengthCm.min())
print('mean:',iris.SepalLengthCm.mean())
```

```python
print("Feature : Sepal Width")
print('max:',iris.SepalWidthCm.max())
print('min:',iris.SepalWidthCm.min())
print('mean:',iris.SepalWidthCm.mean())
```

```python
print("Feature : Petal Length")
print('max:',iris.PetalLengthCm.max())
print('min:',iris.PetalLengthCm.min())
print('mean:',iris.PetalLengthCm.mean())
```

```python
print("Feature : Petal Width")
print('max:',iris.PetalWidthCm.max())
print('min:',iris.PetalWidthCm.min())
print('mean:',iris.PetalWidthCm.mean())
```

```python
import matplotlib.pyplot as plt
```

```python
x = iris.SepalLengthCm
y = iris.PetalLengthCm
plt.scatter(x,y,c='g')
plt.xlabel('Sepal Length(cm)')
plt.ylabel('Petal Length(cm)')
plt.show()
```

```python
plt.bar(x,y)
```

```python
import seaborn as sns
```

```python
x = iris.SepalLengthCm
y = iris.SepalWidthCm
```

```python
sns.jointplot(x,y,iris,color='b')
```

```python
sns.boxplot(iris.Species,iris.SepalWidthCm)
```

```python
sns.boxplot(iris.Species,iris.SepalLengthCm)
sns.stripplot(iris.Species,iris.SepalLengthCm)
```

```python
sns.violinplot(iris.Species,iris.PetalLengthCm,hue=iris.Species)
```

```python
sns.pairplot(iris,hue='Species',vars=['SepalLengthCm', 'SepalWidthCm', 'PetalLengthCm', 'PetalWidthCm'],kind='scatter')
```
